### Question 1(a) (5 marks)

From the data of the tasks.csv, compute a new dataframe called team_memberships. 

Each of its rows contains (user_id, team_id) indicating that the user_id is the assignee of a task owned by the team_id.


```python
""" We are going to exclusively use the csv module built into Python for this task. 
https://www.programiz.com/python-programming/reading-csv-files """
import csv

""" The next thing to do is import the pandas library and give it an alias of pd.  
https://www.w3schools.com/python/pandas/pandas_getting_started.asp """
import pandas as pd

############ Todo 1: ############
## From the data of the tasks.csv, compute a new dataframe called team_memberships. 

""" Read the file using the read cv ( ) method, assuming tabs in the content (i.e. 'It') as separators which will be assigned to the variable df. 

sep - It stands for separator, default is ', ' as in CSV(comma separated values). 
https://www.geeksforgeeks.org/python-read-csv-using-pandas-read_csv/ 

usecols - It is used to retrieve only selected columns from the CSV file. """
team_memberships = pd.read_csv('tasks.csv', usecols = ['assignee', 'team']) #create a dataframe based on tasks.csv and its two columns of 'assignee' and 'team' 


############ Todo 2: ############
## Each of its rows contains (user_id, team_id) indicating that the user_id is the assignee of a task owned by the team_id.

""" Pandas DataFrame.columns attribute return the column labels of the given Dataframe. 
https://www.geeksforgeeks.org/python-pandas-dataframe-columns/

How to rename columns in Pandas DataFrame
Method 2: By assigning a list of new column names 
https://www.geeksforgeeks.org/how-to-rename-columns-in-pandas-dataframe/ """
team_memberships.columns = ['user_id', 'team_id'] #set the columns of the dataframe to be user_id and team_id


########### Todo 3. ###########
""" Sort by columns to see duplicates 
The sort_values() method sorts the DataFrame by the specified label.
https://www.w3schools.com/python/pandas/ref_df_sort_values.asp 

by - Specify labels to sort by. Either index level(s) or column label(s). Or, if the axis is 1 or 'columns' this values specify column level(s) or index label(s) 

inplace - default False. Specifies whether to perform the operation on the original DataFrame or not, if not, which is default, this method returns a new DataFrame """
team_memberships.sort_values(by=['user_id'], inplace=True) #sorts the dataframe by the user_id value


########### Todo 4. ###########
""" The drop_duplicates() method removes duplicate rows.
Use the subset parameter if only some specified columns should be considered when looking for duplicates.
https://www.w3schools.com/python/pandas/ref_df_drop_duplicates.asp 

subset - Optional. A String, or a list, containing the columns to use when looking for duplicates. If not specified, all columns are being used. 

keep - Optional, default 'first'. Specifies which duplicate to keep. If False, drop ALL duplicates

inplace - Optional, default False. If True: the removing is done on the current DataFrame. If False: returns a copy where the removing is done. """
team_memberships.drop_duplicates(subset=None, keep="first", inplace=True) #drops duplicates to clean the data


########### Todo 5. ###########
""" The reset_index() method allows you reset the index back to the default 0, 1, 2 etc indexes.
https://www.w3schools.com/python/pandas/ref_df_reset_index.asp """
team_memberships = team_memberships.reset_index() #resets the index

########### Todo 6. ###########
""" One way to display a dataframe in the form of a table is by using the display() function of IPython.display. 
https://www.geeksforgeeks.org/display-the-pandas-dataframe-in-table-style/ """
display(team_memberships)
```

```python
############ Todo 1: ############
## Add a new column called current_state into the tasks dataframe. Note that the current state of a task can be extracted from the state_transitions dataframe.

""" Read the file using the read cv ( ) method, assuming tabs in the content (i.e. 'It') as separators which will be assigned to the variable df. 

sep - It stands for separator, default is ', ' as in CSV(comma separated values). 
https://www.geeksforgeeks.org/python-read-csv-using-pandas-read_csv/ 

usecols - It is used to retrieve only selected columns from the CSV file. """
current_state = pd.read_csv('state-transitions.csv', usecols = ['task_id', 'state']) #creates a dataframe based on state-transitions.csv file, using the task_id and state columns


########### Todo 2. ###########
""" Pandas DataFrame.columns attribute return the column labels of the given Dataframe. 
https://www.geeksforgeeks.org/python-pandas-dataframe-columns/

How to rename columns in Pandas DataFrame
Method 2: By assigning a list of new column names 
https://www.geeksforgeeks.org/how-to-rename-columns-in-pandas-dataframe/ """
current_state.columns = ['task_id', 'current_state'] #names the columns in the dataframe as task_id and current_state


########### Todo 3. ###########
""" The drop_duplicates() method removes duplicate rows.
Use the subset parameter if only some specified columns should be considered when looking for duplicates.
https://www.w3schools.com/python/pandas/ref_df_drop_duplicates.asp 

subset - Optional. A String, or a list, containing the columns to use when looking for duplicates. If not specified, all columns are being used. 

keep - 'first', 'last', False. Optional, default 'first'. Specifies which duplicate to keep. If False, drop ALL duplicates

inplace - Optional, default False. If True: the removing is done on the current DataFrame. If False: returns a copy where the removing is done. """
current_state.drop_duplicates(subset='task_id', keep="last", inplace=True) #drops the duplicates and keeps the last value - which is the latest state of the task


########### Todo 4. ###########
""" The reset_index() method allows you reset the index back to the default 0, 1, 2 etc indexes.
https://www.w3schools.com/python/pandas/ref_df_reset_index.asp
https://www.geeksforgeeks.org/python-pandas-dataframe-reset_index/
https://www.dataquest.io/blog/tutorial-reset-index-in-pandas/

drop: Boolean value, Adds the replaced index column to the data if False."""
current_state = current_state.reset_index(drop=True)        #resets the index and the replaced index column is completely removed from the DataFrame.


########### Todo 5. ###########
""" Read the file using the read cv ( ) method, assuming tabs in the content (i.e. 'It') as separators which will be assigned to the variable df. 

sep - It stands for separator, default is ', ' as in CSV(comma separated values). 
https://www.geeksforgeeks.org/python-read-csv-using-pandas-read_csv/ 

usecols - It is used to retrieve only selected columns from the CSV file. """
tasks = pd.read_csv('tasks.csv') #creates a dataframe 'tasks' based on tasks.csv


########### Todo 6. ###########
""" Using the 'if not' Python statement to check if it negates the output of an 'if' statement.
https://flexiple.com/python/if-not-python/
https://www.stechies.com/python-not/ 

Pandas DataFrame.columns attribute return the column labels of the given Dataframe. 
https://www.geeksforgeeks.org/python-pandas-dataframe-columns/ """
if not 'current_state' in tasks.columns: #if there isn't a column named 'current_state' in the 'tasks' dataframe, carry out the following codes

    ########### Todo 7. ###########
    """ The merge() method updates the content of two DataFrame by merging them together, using the specified method(s).
    Use the parameters to control which values to keep and which to replace.
    https://www.w3schools.com/python/pandas/ref_df_merge.asp
    https://www.digitalocean.com/community/tutorials/pandas-merge-two-dataframe
    https://www.kdnuggets.com/2023/01/merge-pandas-dataframes.html

    left_on - Specifies in what level to do the merging on the DataFrame to the left

    right_on - Specifies in what level to do the merging on the DataFrame to the right

    how - Default 'inner'. Specifies how to merge 
    left - Left merge uses only the values from the first DataFrame.
    right - Right merge uses only the values from the second DataFrame.
    outer - Include all rows from both DataFrame by using the union of both DataFrame keys.
    inner - Include only the rows that exist in both Dataframes 

    on - Optional. Specifies in what level to do the merging """
    tasks = tasks.merge(current_state, left_on='id', right_on='task_id', how='left') #merges the tasks and current_state dataframes based on their id and task_id column respectively


    ########### Todo 8. ###########
    """ The drop() method removes the specified row or column. 
    https://www.w3schools.com/python/pandas/ref_df_drop.asp 

    axis - 0, 1, 'index', 'columns' - Optional, Which axis to check, default 0. """
    tasks = tasks.drop('task_id', axis=1) #drops the task_id column


    ########### Todo 9. ###########
    """ Here, we simply export a Dataframe to a CSV file using df.to_csv(). 
    https://www.geeksforgeeks.org/saving-a-pandas-dataframe-as-a-csv/ 
    
    sep: the delimiter for the CSV data. It should be a string of length 1, the default is a comma. 
    
    index: if True, index is included in the CSV data. If False, the index value is not written in the CSV output."""
    tasks.to_csv('tasks.csv', sep=',', index=False) # export the 'tasks' Dataframe to the tasks.csv file
    
########### Todo 10. ###########
""" One way to display a dataframe in the form of a table is by using the display() function of IPython.display. 
https://www.geeksforgeeks.org/display-the-pandas-dataframe-in-table-style/ """
display(tasks)
```

The commitment of a team for a sprint is the sum of story points of all tasks assigned to the sprint by the team.

Define a function with the signature compute_commitment(team_id, sprint) which returns the commitment of team indicated by team_id for the sprint indicated by sprint. 

Call the function with (team_id=8, sprint=1) and display the result.

```python
############ Todo 1: ############
## Define a function with the signature compute_commitment(team_id, sprint)

""" We have a function that takes 2 arguments """
def compute_commitment(team_id, sprint): 

    ############ Todo 2: ############
    ## The commitment of a team for a sprint is the sum of story points of all tasks assigned to the sprint by the team.

    """ The groupby() method allows you to group your data and execute functions on these groups.
    https://www.w3schools.com/python/pandas/ref_df_groupby.asp
    https://sparkbyexamples.com/pandas/pandas-groupby-explained-with-examples/ 
    
    The sum() function returns a number, the sum of all items in an iterable.
    https://www.w3schools.com/python/ref_func_sum.asp 
    
    The reset_index() method allows you reset the index back to the default 0, 1, 2 etc indexes.
    https://www.w3schools.com/python/pandas/ref_df_reset_index.asp
    https://www.geeksforgeeks.org/python-pandas-dataframe-reset_index/
    https://www.dataquest.io/blog/tutorial-reset-index-in-pandas/

    drop: Boolean value, Adds the replaced index column to the data if False. """
    commitment = tasks.groupby(['sprint', 'team'])['story_points'].sum().reset_index() #creates a new dataframe based on the tasks dataframe, with the sprint and team columns


    display(commitment)     # commitment Dataframe

    print("\n")


    ############ Todo 3: ############
    ## The function returns the commitment of team indicated by team_id for the sprint indicated by sprint. 

    """ You can then use the loc[ ] method to specify the row using this syntax: 
    The loc property gets, or sets, the value(s) of the specified labels.
    https://www.w3schools.com/python/pandas/ref_df_loc.asp
    https://www.w3resource.com/pandas/dataframe/dataframe-loc.php 
    
    Querying and Filtering:
    Select sprint row in the Dataframe to match the sprint variable. 
    Select team row in the Dataframe to match the team_id variable """
    compute = commitment.loc[(commitment['sprint'] == sprint) & (commitment['team'] == team_id)] #finds the row where the function variables match the dataframe values 

    display(compute)

    print("\n")
    

    ########### Todo 4. ###########
    """ We can select and look at a column, by specifying the name of the column, i.e. 
    https://www.geeksforgeeks.org/dealing-with-rows-and-columns-in-pandas-dataframe/ """
    s_point = compute['story_points']   # Select the 'story_points' column

    display(s_point)

    print("\n")


    ########### Todo 5. ###########
    """ The iloc property gets, or sets, the value(s) of the specified indexes.
    Single index for one row [1] returns a Pandas Series.
    https://www.w3schools.com/python/pandas/ref_df_iloc.asp """ 
    result = s_point.iloc[0] #creates another variable and assigns the story_points in the first row to it

    display(result)

    print("\n")


    ############ Todo 6: ############
    ## display the result.

    """ F-strings provide a concise and convenient way to embed python expressions inside string literals for formatting. 
    https://www.geeksforgeeks.org/formatted-string-literals-f-strings-python/ """
    print(f"The total commitment Team {team_id} has for sprint {sprint} is {result}.")


############ Todo 7: ############
## Call the function with (team_id=8, sprint=1)

""" Keyword arguments (or named arguments) are values that, when passed into a function, are identifiable by specific parameter names.  
https://www.educative.io/answers/what-are-keyword-arguments-in-python """
compute_commitment(team_id = 8, sprint = 1)

```

```python
# ORM library - eliminate the need to deal with underlying raw database instruction

""" We will import sqlalchemy library which is the ORM Python library SQLAlchemy Data Models. """

""" Above, the declarative_base() callable returns a new base class from which all mapped classes should inherit. When the class definition is completed, a new Table and mapper() will have been generated.
https://docs.sqlalchemy.org/en/13/orm/extensions/declarative/basic_use.html """
from sqlalchemy.ext.declarative import declarative_base

""" from the sqlalchemy library we call the various modules
Column - Represents a column in a database table.
Integer - A type for int integers.
String - The base for all string and character types.
Enum - Generic Enum Type.
Float - Type representing floating point types, such as FLOAT or REAL.
ForeignKey - Defines a dependency between two columns. """
from sqlalchemy import Column, Integer, String, Enum, Float, ForeignKey

""" This sqlalchemy library module provides a relationship between two mapped classes. This corresponds to a parent-child or associative table relationship. The constructed class is an instance of RelationshipProperty.
https://docs.sqlalchemy.org/en/14/orm/relationship_api.html """
from sqlalchemy.orm import relationship

""" IMPORT THE SQALCHEMY LIBRARY's CREATE_ENGINE METHOD 
https://www.geeksforgeeks.org/connecting-to-sql-database-using-sqlalchemy-in-python/ """
from sqlalchemy import create_engine


#save the dataframe from 1(d) to csv
########### Todo 1. ###########
""" Here, we simply export a Dataframe to a CSV file using df.to_csv(). 
https://www.geeksforgeeks.org/saving-a-pandas-dataframe-as-a-csv/ 
    
sep: the delimiter for the CSV data. It should be a string of length 1, the default is a comma. 
    
index: if True, index is included in the CSV data. If False, the index value is not written in the CSV output. """
tasks.to_csv('tasks.csv', index=False)


# create a Base object using declarative_base.
########### Todo 2. ###########
""" We need to first create a Base class using declarative_base. This base will be used to create table-classes, from which SQLAlchemy will map it to create the database's table schemas. 

The declarative_base() function is a shorthand version of using the registry.generate_base() method. That is, the following: 
https://docs.sqlalchemy.org/en/14/orm/mapping_api.html#sqlalchemy.orm.declarative_base """
Base = declarative_base()


########### Todo 3. ###########
""" Subsequently, you can use these table-classes to interact with the underlying database through Python codes. SQL libraries require us to define formally the table-schemas (table structure). We will in contrast to other schemaless NoSQL databases which we will look at later. 
"""

""" The following codes define the table, Tasks the same way that the tables are defined in Topic 1: """

# Base object is used to create the databases's table schemas

# define the table Tasks
class Tasks(Base):

    ########### Todo 4. ###########
    """ Table arguments other than the name, metadata, and mapped Column arguments are specified using the __table_args__ class attribute. The __table_args__ class attribute in this case will be assigned a string value.  
    https://docs.sqlalchemy.org/en/13/orm/extensions/declarative/table_config.html """
    __tablename__ = 'tasks'

    ########### Todo 5. ###########
    """ We will use the Column module to create a Column method which will take an integer as first parameter and primary_key attribute would be a True boolean value which would be assigned to a customer_id variable which would be a Column object.   """
    id = Column(Integer, primary_key=True)

    ########### Todo 6. ###########
    """ We will use the Column module to create a Column method which will take the String class as parameter which would be assigned to a variable as a Column object.  """
    description = Column(String)

    ########### Todo 7. ###########
    """ We will use the Column module to create a Column method which will take an integer as first parameter which would be assigned to a variable as a Column object.  """
    assignee = Column(Integer)

    ########### Todo 8. ###########
    """ We will use the Column module to create a Column method which will take an integer as first parameter which would be assigned to a variable as a Column object.  """
    sprint = Column(Integer)

    ########### Todo 9. ###########
    """ We will use the Column module to create a Column method which will take an integer as first parameter which would be assigned to a variable as a Column object.  """
    team = Column(Integer)

    ########### Todo 10. ###########
    """ We will use the Column module to create a Column method which will take an integer as first parameter which would be assigned to a variable as a Column object.  """
    story_points = Column(Integer)

    ########### Todo 11. ###########
    """ We will use the Column module to create a Column method which will take the String class as parameter which would be assigned to a variable as a Column object.  """
    current_state = Column(String)


########### Todo 12. ###########
""" Having declared our Base subclasses in above, we supply its metadata create_all method with our database engine to create our database. """

""" Recall that the basic functions in database operations are to Create, Read, Update and Delete data. When we use an ORM library, we simply manipulate instances of the relevant table-classes.

First, we need to establish a connection with the database, and then start a session to interact with the database. In this example, we will be using the file-based SQLite database. The database file that we are using is order.db: 

We set the echo argument to true so that we can see the output of any SQL instructions generated by SQLAlchemy. Note the use of three backslashes after the colon. 
https://docs.sqlalchemy.org/en/14/core/engines.html """
engine = create_engine('sqlite:///tasks.db', echo=False)


########### Todo 13. ###########
""" create_all() - Create all tables stored in this metadata.
Conditional by default, will not attempt to recreate tables already present in the target database. """
Base.metadata.create_all(engine)


########### Todo 14. ###########
""" Read the file using the read cv ( ) method, assuming tabs in the content (i.e. 'It') as separators which will be assigned to the variable df. 

sep - It stands for separator, default is ', ' as in CSV(comma separated values). 
https://www.geeksforgeeks.org/python-read-csv-using-pandas-read_csv/ 

usecols - It is used to retrieve only selected columns from the CSV file. """
data = pd.read_csv('tasks.csv')


########### Todo 15. ###########
""" Now that you have created a DataFarme, established a connection to a database and also added a table to the database, you can use the Pandas to_sql() function to write the DataFrame into the database.
https://www.askpython.com/python-modules/pandas/pandas-to-sql

name - Name of SQL table

con	- Database connection string

if_exists - Specify what to do if the table already exists in the database
Here, the if_exists='replace' parameter indicates that replace the table if it already exists in the database. """
data.to_sql(name=Tasks.__tablename__, con=engine, if_exists='replace')
```
